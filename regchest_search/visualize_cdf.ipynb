```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
att = pd.read_csv('att_ratio.csv').sort_values('name').rename(columns={'x': 'Validation Error', 'y': 'Cummulative Prob'})
width = pd.read_csv('width_mul.csv').sort_values('name').rename(columns={'x': 'Validation Error', 'y': 'Cummulative Prob'})
stages = pd.read_csv('stages.csv').sort_values('name').rename(columns={'x': 'Validation Error', 'y': 'Cummulative Prob'})
bottleneck = pd.read_csv('bottleneck.csv').sort_values('name').rename(columns={'x': 'Validation Error', 'y': 'Cummulative Prob'})

att = att[att['name'] != 'total']
width = width[width['name'] != 'total']
stages = stages[(stages['name'] != 'total') & (stages['name'] != '-1')]
bottleneck = bottleneck[bottleneck['name'] != 'total']

att = att.rename(columns={'name': 'squeeze ratio s'})
att['squeeze ratio s'] = 's = ' + att['squeeze ratio s']

width = width.rename(columns={'name': 'width multiplier wm'})
width['width multiplier wm'] = 'wm = ' + width['width multiplier wm']

stages = stages.rename(columns={'name': 'stages s'})
stages['stages s'] = 'stage = ' + stages['stages s']

bottleneck = bottleneck.rename(columns={'name': 'bottleneck ratio b'})
bottleneck['bottleneck ratio b'] = 'b = ' + bottleneck['bottleneck ratio b']
```

```python
plt.rcParams["axes.labelsize"] = 35
_, axes = plt.subplots(ncols=4, figsize=(50, 10))

sns.lineplot(data=att, x='Validation Error', y='Cummulative Prob', style='squeeze ratio s', hue='squeeze ratio s', linewidth=5, ax=axes[0])
axes[0].set_title('a. Squeeze ratio', y=-0.25, fontsize=35)
axes[0].legend(fontsize=24, loc="lower right")
axes[0].xaxis.set_tick_params(labelsize=30)
axes[0].yaxis.set_tick_params(labelsize=30)

sns.lineplot(data=width, x='Validation Error', y='Cummulative Prob', style='width multiplier wm', hue='width multiplier wm', linewidth=5, ax=axes[1])
axes[1].set_title('b. Width multiplier', y=-0.25, fontsize=35)
axes[1].legend(fontsize=24)
axes[1].xaxis.set_tick_params(labelsize=30)
axes[1].yaxis.set_tick_params(labelsize=30)

sns.lineplot(data=stages, x='Validation Error', y='Cummulative Prob', style='stages s', hue='stages s', linewidth=5, ax=axes[2])
axes[2].set_title('c. Stages', y=-0.25, fontsize=35)
axes[2].legend(fontsize=24, loc="lower right")
axes[2].xaxis.set_tick_params(labelsize=30)
axes[2].yaxis.set_tick_params(labelsize=30)

sns.lineplot(data=bottleneck, x='Validation Error', y='Cummulative Prob', style='bottleneck ratio b', hue='bottleneck ratio b', linewidth=5, ax=axes[3])
axes[3].set_title('d. Bottleneck ratio', y=-0.25, fontsize=35)
axes[3].legend(fontsize=24, loc="lower right")
axes[3].xaxis.set_tick_params(labelsize=30)
axes[3].yaxis.set_tick_params(labelsize=30)

plt.show()
```
